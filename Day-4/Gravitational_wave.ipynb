```python
import numpy as np
import matplotlib.pyplot as plt
strain, freq, time = np.loadtxt('GW_data_file.csv',delimiter=',')
plt.figure(1)
plt.plot(time,strain,'r-')
plt.xlabel('Time(t)')
plt.ylabel('Strain data')
plt.title('Strain vs time plot')
plt.grid()
plt.figure(2)
plt.plot(time,freq,'g+',marker=',')
plt.xlabel('Time(t)')
plt.ylabel('Freq')
plt.title('Frequency vs time plot')
plt.grid()
zero_crossings = np.where(np.diff(np.sign(strain)))[0]
time=time[zero_crossings]
freq=freq[zero_crossings]
plt.figure(3)
plt.text(-0.175,100,'inspiral')
plt.scatter(time[:38],freq[:38],marker='2')
plt.text(-0.025,200,'merger')
plt.scatter(time[38:47],freq[38:47],marker='3')
plt.text(0.030,120,'ringdown',horizontalalignment='right')
plt.scatter(time[47:],freq[47:],marker='1')
plt.xlabel('Time(t)')
plt.ylabel('Freq')
plt.title('Frequency vs time plot')
plt.grid()
```
